# Example: REV analysis for porosity

This example shows the functionality of REVAnalyzer library and the structure of output for vector metric REV
analysis taking correla as the example. 

```python
from julia.api import Julia
jl = Julia(compiled_modules=False)
#These two lines are necessary in every code which uses REVAnalyzer library, even if this code 
#does not use julia packages. Error is generarted if you skip these line in your code.  
```

image = 'ceramic300'
size = 300
inputdir = 'data'
cut_step = 50
sREV_max_size = 200

```python
vectorizer = CFVectorizer(mode='max')
metric = S2(vectorizer, show_time=True, normalize = True)
dREV_threshold = 0.2
sREV_threshold = 0.2
analyzer = REVAnalyzer(metric, image, size, cut_step, sREV_max_size, dREV_threshold, sREV_threshold)
```

```python
analyzer.generate()
```

```python
cut_size = 150
cut_id = 5
value = analyzer.read(cut_size, cut_id)
print(value)
```

```python
metric.show(analyzer.outputdir_cuts_values, image, cut_size, cut_id)
```

```python
analyzer.vectorize()
```

```python
analyzer.analyze()
print(analyzer.metric_mean)
print(analyzer.metric_std)
print(analyzer.metric_normed_std)
print(analyzer.metric_normed_std_1)
print(analyzer.metric_normed_std_2)
print(analyzer.dREV_size_1)
print(analyzer.sREV_size_1)
print(analyzer.sREV_size_2)
```

```python
analyzer.show_results()
```

```python
vectorizer = CFVectorizer(mode='all')
metric = S2(vectorizer, show_time=True)
dREV_threshold = 0.2
sREV_threshold = 0.2
analyzer = REVAnalyzer(metric, image, size, cut_step, sREV_max_size, dREV_threshold, sREV_threshold)
analyzer.vectorize()
analyzer.analyze()
print(analyzer.metric_mean)
print(analyzer.metric_std)
print(analyzer.metric_normed_std)
print(analyzer.metric_normed_std_1)
print(analyzer.metric_normed_std_2)
print(analyzer.dREV_size_1)
print(analyzer.sREV_size_1)
print(analyzer.sREV_size_2)
```

```python
analyzer.show_results()
```
